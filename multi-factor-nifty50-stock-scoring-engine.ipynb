```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("/kaggle/input/nifty50-factors/nifty50_factors.csv")
df = df.dropna()
df['score'] = df['score'].astype(float)
df['pe'] = df['pe'].astype(float)
df['changePercent'] = df['changePercent'].astype(float)

# PE vs Change Visualization
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='pe', y='changePercent', hue='score', size='marketCap', palette='viridis')
plt.title("PE vs % Change (%) - Sized by Market Cap")
plt.xlabel("Price-to-Earnings (P/E)")
plt.ylabel("Daily % Change")
plt.grid(True)
plt.show()

# Top 10 by score
top10 = df.sort_values(by='score', ascending=False).head(10)
top10[['symbol', 'lastPrice', 'changePercent', 'pe', 'score']]

# Bar plot
plt.figure(figsize=(10,5))
sns.barplot(data=top10, x='symbol', y='score', palette='coolwarm')
plt.title("Top 10 Stocks by Multi-Factor Score")
plt.ylabel("Score")
plt.xticks(rotation=45)
plt.show()
```
